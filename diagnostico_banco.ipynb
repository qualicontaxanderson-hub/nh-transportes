## 3. FUNÇÕES AUXILIARES

````python
def execute_query(query, params=None):
    """Executa query e retorna DataFrame"""
    conn = pymysql.connect(**DB_CONFIG)
    try:
        df = pd.read_sql(query, conn, params=params)
        return df
    except Exception as e:
        print(f"❌ Erro: {e}")
        return None
    finally:
        conn.close()

def print_section(title):
    """Imprime título de seção"""
    print("\n" + "="*80)
    print(f"  {title}")
    print("="*80 + "\n")

def print_result(df, title):
    """Imprime resultado formatado"""
    print(f"\n📊 {title}")
    print("-" * 80)
    if df is not None and not df.empty:
        print(df.to_string(index=False))
        print(f"\nTotal de registros: {len(df)}")
    else:
        print("⚠️ Nenhum resultado encontrado.")
    print("-" * 80)

print("✅ Funções auxiliares carregadas!")
````

## 4. DIAGNÓSTICO 1: ESTRUTURA DA TABELA FUNCIONARIOS

````python
print_section("DIAGNÓSTICO 1: ESTRUTURA DA TABELA FUNCIONARIOS")

query = "SHOW COLUMNS FROM funcionarios"
df = execute_query(query)

if df is not None:
    print_result(df, "Estrutura da Tabela funcionarios")
    
    # Verificar se coluna categoria existe
    if 'categoria' in df['Field'].values:
        print("\n✅ COLUNA 'categoria' EXISTE!")
        cat_info = df[df['Field'] == 'categoria'].iloc[0]
        print(f"   - Tipo: {cat_info['Type']}")
        print(f"   - Permite NULL: {cat_info['Null']}")
    else:
        print("\n❌ COLUNA 'categoria' NÃO EXISTE!")
        print("   Solução: ALTER TABLE funcionarios ADD COLUMN categoria VARCHAR(50);")
````

## 5. DIAGNÓSTICO 2: DADOS DA TABELA FUNCIONARIOS

````python
print_section("DIAGNÓSTICO 2: DADOS DA TABELA FUNCIONARIOS")

query = """
SELECT 
    id,
    nome,
    categoria,
    CASE 
        WHEN categoria IS NULL THEN '⚠️ NULL'
        WHEN categoria = '' THEN '⚠️ VAZIO'
        ELSE '✅ OK'
    END as status_categoria
FROM funcionarios
ORDER BY id
"""

df = execute_query(query)
print_result(df, "Funcionários Cadastrados")

if df is not None:
    # Análise
    total = len(df)
    nulls = len(df[df['status_categoria'] == '⚠️ NULL'])
    vazios = len(df[df['status_categoria'] == '⚠️ VAZIO'])
    ok = len(df[df['status_categoria'] == '✅ OK'])
    
    print("\n📊 ANÁLISE:")
    print(f"   Total de funcionários: {total}")
    print(f"   Com categoria OK: {ok} ✅")
    print(f"   Com categoria NULL: {nulls} {'❌' if nulls > 0 else '✅'}")
    print(f"   Com categoria VAZIA: {vazios} {'❌' if vazios > 0 else '✅'}")
    
    if nulls > 0 or vazios > 0:
        print("\n⚠️ PROBLEMA ENCONTRADO: Categoria NULL ou vazia!")
        print("   Solução: UPDATE funcionarios SET categoria = 'FRENTISTA' WHERE categoria IS NULL;")
    else:
        print("\n✅ Todas as categorias estão preenchidas!")
    
    # Valores distintos
    if 'categoria' in df.columns:
        categorias = df['categoria'].dropna().unique()
        print(f"\n   Valores de categoria: {list(categorias)}")
````

## 6. DIAGNÓSTICO 3: DADOS DA TABELA MOTORISTAS

## 7. DIAGNÓSTICO 4: SOBREPOSIÇÃO DE IDs

````python
print_section("DIAGNÓSTICO 4: SOBREPOSIÇÃO DE IDs")

query = """
SELECT 
    f.id as func_id,
    f.nome as func_nome,
    f.categoria as func_categoria,
    m.id as mot_id,
    m.nome as mot_nome,
    CASE 
        WHEN f.id IS NOT NULL AND m.id IS NOT NULL THEN '⚠️ AMBAS'
        WHEN f.id IS NOT NULL THEN '✅ SÓ FUNC'
        WHEN m.id IS NOT NULL THEN '✅ SÓ MOT'
    END as status
FROM funcionarios f
LEFT JOIN motoristas m ON f.id = m.id

UNION

SELECT 
    f.id as func_id,
    f.nome as func_nome,
    f.categoria as func_categoria,
    m.id as mot_id,
    m.nome as mot_nome,
    CASE 
        WHEN f.id IS NOT NULL AND m.id IS NOT NULL THEN '⚠️ AMBAS'
        WHEN f.id IS NOT NULL THEN '✅ SÓ FUNC'
        WHEN m.id IS NOT NULL THEN '✅ SÓ MOT'
    END as status
FROM motoristas m
LEFT JOIN funcionarios f ON m.id = f.id
WHERE f.id IS NULL

ORDER BY COALESCE(func_id, mot_id)
"""

df = execute_query(query)
print_result(df, "Análise de Sobreposição de IDs")

if df is not None:
    ambas = len(df[df['status'] == '⚠️ AMBAS'])
    so_func = len(df[df['status'] == '✅ SÓ FUNC'])
    so_mot = len(df[df['status'] == '✅ SÓ MOT'])
    
    print("\n📊 ANÁLISE:")
    print(f"   IDs em ambas tabelas: {ambas} {'⚠️' if ambas > 0 else '✅'}")
    print(f"   IDs só em funcionarios: {so_func}")
    print(f"   IDs só em motoristas: {so_mot}")
    
    if ambas > 0:
        print("\n⚠️ IDs SOBREPOSTOS ENCONTRADOS!")
        print("   Mas a query com COALESCE trata isso corretamente.")
        print("   Funcionários têm prioridade sobre motoristas.")
        print("\n   IDs sobrepostos:")
        sobrepostos = df[df['status'] == '⚠️ AMBAS'][['func_id', 'func_nome', 'mot_nome']]
        print(sobrepostos.to_string(index=False))
````

## 8. DIAGNÓSTICO 5: LANÇAMENTOS DO MÊS

````python
print_section("DIAGNÓSTICO 5: LANÇAMENTOS DO MÊS 01/2026")

# Parâmetros (ajustar conforme necessário)
MES = '01/2026'
CLIENTE_ID = 1

query = """
SELECT 
    l.funcionarioid,
    f.nome as func_nome,
    f.categoria as func_categoria,
    m.nome as mot_nome,
    COUNT(*) as qtd_lancamentos,
    SUM(l.valor) as total_valor,
    CASE 
        WHEN f.id IS NOT NULL THEN COALESCE(f.categoria, 'FRENTISTAS')
        WHEN m.id IS NOT NULL THEN 'MOTORISTAS'
        ELSE 'OUTROS'
    END as categoria_calculada
FROM lancamentosfuncionarios_v2 l
LEFT JOIN funcionarios f ON l.funcionarioid = f.id
LEFT JOIN motoristas m ON l.funcionarioid = m.id
WHERE l.mes = %s AND l.clienteid = %s
GROUP BY l.funcionarioid, f.nome, f.categoria, m.nome
ORDER BY l.funcionarioid
"""

df = execute_query(query, params=(MES, CLIENTE_ID))
print_result(df, f"Lançamentos de {MES} - Cliente {CLIENTE_ID}")

if df is not None:
    print("\n📊 ANÁLISE POR CATEGORIA:")
    categorias = df.groupby('categoria_calculada').agg({
        'funcionarioid': 'count',
        'total_valor': 'sum'
    }).reset_index()
    categorias.columns = ['Categoria', 'Total Funcionários', 'Valor Total']
    print(categorias.to_string(index=False))
    
    print("\n📊 COMPARAÇÃO COM ESPERADO:")
    esperado = {
        'FRENTISTAS': (7, 23263.98),
        'FRENTISTA': (7, 23263.98),
        'MOTORISTAS': (2, 6308.45)
    }
    
    for _, row in categorias.iterrows():
        cat = row['Categoria']
        qtd = row['Total Funcionários']
        
        if cat in esperado:
            exp_qtd, exp_val = esperado[cat]
            status = '✅' if qtd == exp_qtd else '❌'
            print(f"   {cat}: {qtd} funcionários {status} (esperado: {exp_qtd})")
        else:
            print(f"   {cat}: {qtd} funcionários ⚠️ (categoria inesperada)")
````

## 9. DIAGNÓSTICO 6: SIMULAÇÃO DA QUERY DO SISTEMA

````python
print_section("DIAGNÓSTICO 6: SIMULAÇÃO DA QUERY DO SISTEMA")

print("Esta é a query EXATA que o sistema usa após o commit 75:")
print()

query = """
SELECT 
n    sub.mes,
    sub.cliente_nome,
    sub.categoria,
    COUNT(DISTINCT sub.funcionarioid) as total_funcionarios,
    SUM(sub.valor) as total_valor,
    sub.statuslancamento
FROM (
    SELECT 
        l.mes,
        c.razao_social as cliente_nome,
        l.funcionarioid,
        CASE 
            WHEN f.id IS NOT NULL THEN COALESCE(f.categoria, 'FRENTISTAS')
            WHEN m.id IS NOT NULL THEN 'MOTORISTAS'
            ELSE 'OUTROS'
        END as categoria,
        l.valor,
        l.statuslancamento
    FROM lancamentosfuncionarios_v2 l
    LEFT JOIN clientes c ON l.clienteid = c.id
    LEFT JOIN funcionarios f ON l.funcionarioid = f.id
    LEFT JOIN motoristas m ON l.funcionarioid = m.id
    WHERE l.mes = %s AND l.clienteid = %s
) as sub
GROUP BY sub.mes, sub.cliente_nome, sub.categoria, sub.statuslancamento
ORDER BY sub.mes DESC, sub.categoria
"""

df = execute_query(query, params=(MES, CLIENTE_ID))
print_result(df, "RESULTADO DA QUERY DO SISTEMA")

if df is not None:
    print("\n" + "="*80)
    print("  🎯 RESULTADO FINAL")
    print("="*80)
    
    frentistas = df[df['categoria'].str.contains('FRENTISTA', case=False, na=False)]
    motoristas = df[df['categoria'] == 'MOTORISTAS']
    
    print("\n✅ FRENTISTAS:")
    if not frentistas.empty:
        for _, row in frentistas.iterrows():
            status = '✅' if row['total_funcionarios'] == 7 else '❌'
            print(f"   {row['categoria']}: {row['total_funcionarios']} funcionários {status}")
            print(f"   Valor: R$ {row['total_valor']:,.2f}")
    else:
        print("   ❌ Nenhum frentista encontrado!")
    
    print("\n✅ MOTORISTAS:")
    if not motoristas.empty:
        for _, row in motoristas.iterrows():
            status = '✅' if row['total_funcionarios'] == 2 else '❌'
            print(f"   MOTORISTAS: {row['total_funcionarios']} funcionários {status}")
            print(f"   Valor: R$ {row['total_valor']:,.2f}")
    else:
        print("   ❌ Nenhum motorista encontrado!")
    
    # Verificação final
    total_func = df['total_funcionarios'].sum()
    print(f"\n📊 TOTAL: {total_func} funcionários")
    
    if total_func == 9:
        print("   ✅ Total correto!")
    else:
        print(f"   ❌ Total incorreto! Esperado: 9, Encontrado: {total_func}")
````
